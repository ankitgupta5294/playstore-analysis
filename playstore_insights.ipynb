<h2> Define Logical Combinations

```python
logical_combinations = [
    ["free"],
    ["genre"],
    ["releasedYear"],
    ["free", "genre"],
    ["free", "minInstalls"],
    ["free", "ratings", "reviews", "score"],
    ["price", "ratings"],
    ["len screenshots", "ratings"],
    ["free", "reviews"],
    ["free", "score"],
    ["offersIAP", "adSupported", "containsAds"],
    ["offersIAP", "score"],
    ["offersIAP", "minInstalls"],
    ["releasedYear", "offersIAP"],
    ["containsAds", "ratings"]
]
```

```python
def output_formatter(input_dict, columns):
    output_str = ""
    for key, value in input_dict.items():
        if key in columns:
            output_str += f"{key}={value}; "
        elif key == 'count':
            output_str = output_str[:-2] + f", {value}\n"
    return output_str
```

```python
with open("results.csv", "w") as output_file:
    for combination in logical_combinations:
        for i in pd.DataFrame(df_final[combination].value_counts()).reset_index().to_dict(orient="records"):
            output_file.write(output_formatter(i, combination))
```
